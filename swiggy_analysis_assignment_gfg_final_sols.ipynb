```python
def cost_fix(string):
  if type(string)==str:
    string=string.replace('₹','')
    string=string.strip()
    string=int(string)
  return string
```

```python
for i in df['city']:
  i=i.split(',')
  if len(i)==1:
    i=i[0]
  elif len(i)>1:
    i=i[-1]

print(i)
```

```python
def rating_fix(string):
  if string=='--':
    string=0.0
  else:
    string=float(string)

  return string
```

```python
df['rating']=df['rating'].apply(lambda string : rating_fix(string))
df['cost']=df['cost'].apply(lambda string : cost_fix(string))
```

```python
query='''
SELECT city , COUNT(DISTINCT(name)) as restaurant_count FROM df
GROUP BY city
ORDER BY restaurant_count DESC
LIMIT 10

'''

sqldf(query,globals())
```

Most Popular Cuisines served throughout the dataset?


```python
df.columns
```

```python
query='''
SELECT cuisine , COUNT(cuisine) as cuisine_popularity FROM df
GROUP BY cuisine
ORDER BY cuisine_popularity DESC
LIMIT 1
'''

sqldf(query,globals())
```

Which city is having the least expensive restaurant in terms of cost?


```python
query='''
SELECT city , MIN(cost) as restaurant_expense FROM df
GROUP BY city
ORDER BY restaurant_expense ASC
LIMIT 1
'''

sqldf(query,globals())
```

Top 5 most popular restaurant chains in India?

```python
query='''
SELECT name , COUNT(name) as branch_count FROM df
GROUP BY name
ORDER BY branch_count DESC
LIMIT 5

'''

sqldf(query,globals())
```

Which restaurant in Banglore has the most number of people visited(take a number of reviews for reference)?


```python
query='''
SELECT name , COUNT(rating) as popularity FROM df
WHERE city LIKE '%Bangalore%'
GROUP BY name
ORDER BY popularity DESC
LIMIT 1

'''

sqldf(query,globals())
```
